```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
OrderList = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name='OrderList')
FreightRates = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name='FreightRates')
WhCosts = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name='WhCosts')
WhCapacities = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name='WhCapacities')
ProductsPerPlant = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name='ProductsPerPlant')
PlantPorts = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name='PlantPorts')
```

```python
OrderList['Origin_Port'] = OrderList['Origin_Port'].str.replace('PORT', '')
OrderList['Origin_Port'] = OrderList['Origin_Port'].astype(int)
```

```python
OrderList['Plant_Code'] = OrderList['Plant_Code'].str.replace('PLANT', '')
OrderList['Plant_Code'] = OrderList['Plant_Code'].astype(int)

OrderList['Destination_Port'] = OrderList['Destination_Port'].str.replace('PORT', '')
OrderList['Destination_Port'] = OrderList['Destination_Port'].astype(int)
```

```python
# make sure all demanded products are produced by plants
print(OrderList.shape)

IDs = ProductsPerPlant['Product_ID']
IDs = np.array(IDs)

to_remove = []

for i in range(len(OrderList['Product_ID'])):
    if OrderList['Product_ID'].iloc[i] not in IDs:
        to_remove.append(i)

OrderList.drop(to_remove, inplace=True)

print(OrderList.shape)
```

```python
encoder = LabelEncoder()
ProductsPerPlant['Product_ID'] = encoder.fit_transform(ProductsPerPlant['Product_ID'])
OrderList['Product_ID'] = encoder.transform(OrderList['Product_ID'])
```

```python
# drop products not demanded

print(ProductsPerPlant.shape)

IDs = OrderList['Product_ID']
IDs = np.array(IDs)

to_remove = []

for i in range(len(ProductsPerPlant['Product_ID'])):
    if ProductsPerPlant['Product_ID'].iloc[i] not in IDs:
        to_remove.append(i)

ProductsPerPlant.drop(to_remove, inplace=True)
ProductsPerPlant = ProductsPerPlant.reset_index(drop=True)
print(ProductsPerPlant.shape)
```

```python
# check if all ordered items are produced at any plant

IDs = ProductsPerPlant['Product_ID']
IDs = np.array(IDs)

for i in range(len(OrderList['Product_ID'])):
    if OrderList['Product_ID'].iloc[i] not in IDs:
        print("oops")
```
